#Задание 1

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
_font = {
    "fontsize": 10,
    "family": "arial",}
plt.plot(x, y)
plt.xlabel("Переменная x", fontdict=_font)
plt.ylabel("Переменная y", fontdict=_font)
plt.show()
```

```python
plt.scatter(x, y, marker='o', c='r', edgecolor='b')
plt.title("Диаграмма разброса")
plt.xlabel("Переменная x", fontdict=_font)
plt.ylabel("Переменная y", fontdict=_font)
plt.show()
```

#Задание 2

```python
t = np.linspace(0, 10, num=51)
f = np.cos(t)
```

```python
plt.plot(t, f, color='g')
plt.title("График f(t)")
plt.xlabel("Значения t", fontdict=_font)
plt.ylabel("Значения f", fontdict=_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

#Задание 3

```python
x = np.linspace(-3, 3, num=51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График y1')
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
```
